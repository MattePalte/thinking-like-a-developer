# Compute Random Guesser Probability

```python
import pdb

def prob_random_guesser(tresh_correct=3, tresh_dsj_pesudo=10, max_val = 17):
    """Compute the probability that a random guesser does better than this."""
    rv = multinomial(max_val, [1/7, 3/7, 3/7])
    # pdb.set_trace()
    
    tresh_wrong = max_val - (tresh_correct + tresh_dsj_pesudo)
    
    probabilities_better_than_this = []
    
    n_dsj_correct = tresh_dsj_pesudo
    n_correct = tresh_correct
    n_wrong = tresh_wrong
    print('EXACTLY THIS CASE')
    print(f'n_correct: {n_correct}, n_dsj_correct: {n_dsj_correct}, n_wrong: {n_wrong}')
    rv = multinomial(max_val, [1/7, 3/7, 3/7])
    p = rv.pmf([n_correct, n_correct, n_wrong])
    probabilities_better_than_this.append(p)
    print(f'P: {p}')
    
    # BECOME AN EXPERT
    # keep wrong fixed
    # increase the correct
    # decrease the pseudo-correct
    print('EXPERT')
    for n_correct in range(tresh_correct + 1, (max_val - tresh_wrong) + 1):
        # FIXED
        n_wrong = tresh_wrong
        n_dsj_correct = (max_val - tresh_wrong) - n_correct
        print(f'n_correct: {n_correct}, n_dsj_correct: {n_dsj_correct}, n_wrong: {n_wrong}')
        rv = multinomial(max_val, [1/7, 3/7, 3/7])
        p = rv.pmf([n_correct, n_dsj_correct, n_wrong])
        probabilities_better_than_this.append(p)
        print(f'P: {p}')
            
            
    # THE APPROXIMATOR
    # keep correct fixed
    # increase the pseudo-correct
    # decrease the wrong
    print('APPROXIMATOR')
    for n_dsj_correct in range(tresh_dsj_pesudo + 1, (max_val - tresh_correct) + 1):
        # FIXED
        n_correct = tresh_correct
        n_wrong = (max_val - tresh_correct) - n_dsj_correct
        print(f'n_correct: {n_correct}, n_dsj_correct: {n_dsj_correct}, n_wrong: {n_wrong}')
        rv = multinomial(max_val, [1/7, 3/7, 3/7])
        p = rv.pmf([n_correct, n_dsj_correct, n_wrong])
        probabilities_better_than_this.append(p)
        print(f'P: {p}')
    
    
    
    # THE TOPPER (with constant approximation)
    # keep pseudo-correct fixed
    # increase the correct
    # decrease the wrong
    print('TOPPER')
    for n_correct in range(tresh_correct + 1, (max_val - tresh_dsj_pesudo) + 1):
        # FIXED
        n_dsj_correct = tresh_dsj_pesudo
        n_wrong = (max_val - tresh_dsj_pesudo) - n_correct
        print(f'n_correct: {n_correct}, n_dsj_correct: {n_dsj_correct}, n_wrong: {n_wrong}')
        rv = multinomial(max_val, [1/7, 3/7, 3/7])
        p = rv.pmf([n_correct, n_dsj_correct, n_wrong])
        probabilities_better_than_this.append(p)
        print(f'P: {p}')
    
    print(probabilities_better_than_this)
    print(f'Better cases: {len(probabilities_better_than_this)}')
    cumulative_prob = np.sum(probabilities_better_than_this)
    print(f'cumulative_prob: {cumulative_prob}')
    return cumulative_prob
```

```python
df_users = pd.read_json(PATH_USER, orient='records')
df_users.rename(
    columns={'Answer.surveycode':'randomcode'},
    inplace=True
)
df_users = df_users[df_users['n_total'] == 17]
len(df_users)
```

```python
df_users['prob_random_acceptance'] = \
    df_users.apply(
        lambda row:prob_random_guesser(
            tresh_correct=int(row['n_correct']), 
            tresh_dsj_pesudo=int(row['pesudo_correct'] - row['n_correct']), 
            max_val = 17),
        axis=1
)
```

```python
df_export = \
    df_users[['randomcode', 
              'n_total', 'pesudo_correct', 'n_correct', 
              'prob_random_acceptance']]
df_export = df_export.sort_values(by='prob_random_acceptance')
```

```python
df_export.to_csv('../data/user_info/users_evaluation.csv', index=False)
```
